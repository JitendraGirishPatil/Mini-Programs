## Description
### A Shopping Cart with following basic operations :
 - Displays a menu with items wiht corresponding quantities available in store and cost of the item.
 - Accepts customer input for items required.
 - Displays a message if the item is out of stock .
 - Takes input of customer details.
 - Calculates delivery charges based on the distance.
 - Displays a final bill with the items purchased and the customer details.


```python
cost_pep = 35.5
cost_bisc = 40
cost_snacks = 55.5
cost_chips = 30.5
cost_choco = 100

pep = 40
bisc = 30
snacks = 10
chips = 5
choco = 8

def add():
    print()
    print()
    print('***************************** WELCOME TO SUPER MARKET SHOP ******************************')
    print()
    print()
    print('%s %20s %10s %10s' %('S.No','Name of Product','Cost/unit','Quantity'))
    print('%d %20s %12.2f %10d' %(1,'Pepsodent',cost_pep,pep))
    print('%d %20s %12.2f %10d' %(2,'Biscuit',cost_bisc,bisc))
    print('%d %20s %12.2f %10d' %(3,'Snacks',cost_snacks,snacks))
    print('%d %20s %12.2f %10d' %(4,'Chips',cost_chips,chips))
    print('%d %20s %12.2f %10d' %(5,'Chocolate',cost_choco,choco))
    print()
    
    qty_pep = int(input('Enter Quantity of Pepsodent : '))
    if (pep<qty_pep):
        print()
        print(' Sorry !! This item is in low stock than you want. Available stock is :',pep)
        qty_pep = int(input('Enter Quantity of Pepsodent again : '))
    elif (pep==0):
        print(' Sorry !! This item is out of stock !!! ')
    
        
    qty_bisc = int(input('Enter Quantity of Biscuits : '))
    if (bisc<qty_bisc):
        print()
        print(' Sorry !! This item is in low stock than you want. Available stock is :',bisc)
        qty_bisc = int(input('Enter Quantity of Biscuits again : '))
    elif (bisc==0):
        print(' Sorry !! This item is out of stock !!! ')
    
        
    qty_snacks = int(input('Enter Quantity of Snacks : '))
    if (snacks<qty_snacks):
        print()
        print(' Sorry !! This item is in low stock than you want. Available stock is :',snacks)
        qty_snacks = int(input('Enter Quantity of Snacks again : '))
    elif (snacks==0):
        print(' Sorry !! This item is out of stock !!! ')
    
        
    qty_chips = int(input('Enter Quantity of Chips : '))
    if (chips<qty_chips):
        print()
        print(' Sorry !! This item is in low stock than you want. Available stock is :',chips)
        qty_chips = int(input('Enter Quantity of Chips again : '))
    elif (chips==0):
        print(' Sorry !! This item is out of stock !!! ')
    
        
    qty_choco = int(input('Enter Quantity of Chocolate : '))
    if (choco<qty_choco):
        print()
        print(' Sorry !! This item is in low stock than you want. Available stock is :',choco)
        qty_choco = int(input('Enter Quantity of Chocolate again : '))
    elif (choco==0):
        print(' Sorry !! This item is out of stock !!! ')
        print()    
    
    return qty_pep,qty_bisc,qty_snacks,qty_chips,qty_choco

def tot_cost(qty_pep,qty_bisc,qty_snacks,qty_chips,qty_choco):
    total_cost = qty_pep*cost_pep + qty_bisc*cost_bisc + qty_snacks*cost_snacks + qty_chips*cost_chips + qty_choco*cost_choco
    return total_cost

def deli_details():
    print('Please Enter your Delivery details')
    print()
    name = input('Enter the name : ')
    address = input('Enter the address : ')
    dist = float(input('Enter the distance of your house from the store : '))
    print()
    
    return name,address,dist

def cal_deli_charge(dist):
    if dist<=5 :
        deli_charge=0
    else:
        deli_charge=100
    return deli_charge

def calc_final_bill(total_cost,deli_charge):
    final_bill = total_cost + deli_charge
    return final_bill

def bill(qty_pep,qty_bisc,qty_snacks,qty_chips,qty_choco,name,address,total_cost,final_bill):
    print()
    print('************** Dear Customer YOUR FINAL LIST ******************\n--------------------------------------------------------------')
    print()
    print('%s %22s %11s %7s %15s' %('S.No','Name of Product','Cost/unit','Quantity','Total Cost'))
    print('%d %20s %12d %10.2f %15.2f' %(1,'Pepsodent',cost_pep,qty_pep,qty_pep*cost_pep))
    print('%d %20s %12d %10.2f %15.2f' %(2,'Biscuit',cost_bisc,qty_bisc,qty_bisc*cost_bisc))
    print('%d %20s %12d %10.2f %15.2f' %(3,'Snacks',cost_snacks,qty_snacks,qty_snacks*cost_snacks))
    print('%d %20s %12d %10.2f %15.2f' %(4,'Chips',cost_chips,qty_chips,qty_chips*cost_chips))
    print('%d %20s %12d %10.2f %15.2f' %(5,'Chocolate',cost_choco,qty_choco,qty_choco*cost_choco))
    print()
    print('*****************************************************************')
    print()
    print('Name :', total_cost)
    print('Address: ', final_bill)
    print('Total cost of item : ',name)
    print('Total Bill including Delivery charges is : ',address)
    
    
    
while True:
    qty_pep,qty_bisc,qty_snacks,qty_chips,qty_choco = add()
    total_cost= tot_cost(qty_pep,qty_bisc,qty_snacks,qty_chips,qty_choco)
    
    pep = pep - qty_pep 
    bisc = bisc - qty_bisc 
    snacks = snacks - qty_snacks 
    chips = chips - qty_chips 
    choco = choco - qty_choco
    
    print()
    print()
    name,address,dist = deli_details()
    deli_charge = cal_deli_charge(dist)
    final_bill = calc_final_bill(total_cost,deli_charge)
    bill(qty_pep,qty_bisc,qty_snacks,qty_chips,qty_choco,total_cost,final_bill,name,address)
    print()
    choice = input('Would you like to continue shopping with us ? (Y/N)')
    print()
    if (choice== 'Y' ) or (choice == 'y' ):
        continue
    else:
        print('-*-*-*-*-*-*-*- Thank You for shopping with us, Have a great day -*-*-*-*-*-*-*-')
        break
```
