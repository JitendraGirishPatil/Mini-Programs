```python
import math

def rad_to_deg(rad):
    import math
    rad = float(input('Enter the Radian to be converted : '))
    math.degrees(rad)
    print('The Degree of',rad,'is :',math.degrees(rad),'°')
    return math.degrees(rad)

def deg_to_rad(deg):
    import math
    deg = float(input('Enter the Degree to be converted : '))
    math.radians(deg)
    print('The Radian of',deg,'is :',math.radians(deg),'rad')
    return math.radians(deg)

def fact(num):
    import math
    num = int(input('Enter the number : '))
    print('Factorial of ',num,'is :',math.factorial(num))
    return math.factorial(num)

def sqrt(num):
    import math
    num = int(input('Enter the number : '))
    print('Square root of ',num,'is :',math.sqrt(num))
    return math.sqrt(num)
    
def mb_to_n(num):
    num = int(input('Enter the number : '))
    n=num*2000000
    print('Conversion of mb to nibble ',num,'is :',n,'n')
    return n
    
def km_to_ms(num):
    num = float(input('Enter the number : '))
    ms = num*5/18
    print('Conversion from km/hr to m/s is :',ms,'ms')
    return ms

def ms_to_km(num):
    num = float(input('Enter the number : '))
    km = 18*num/5
    print('Conversion from km/hr to m/s is :',km,'km')
    return km
    
def int_to_flo(num):
    num = int(input('Enter the number : '))
    a = float(num)
    print('Integer to Float value is :',a)
    return a
    
def flo_to_int(num):
    num = float(input('Enter the number : '))
    b = int(num)
    print('Integer to Float value is :',b)
    return b
    
def dec_to_oct(num):
    num = int(input('Enter the number : '))
    o = oct(num)
    print('Decimal to Octal Value is :',o)
    return o
    
def oct_to_dec(num):
    num = (input('Enter the number : '))
    int(num,8)
    print('Octal to Decimal Value is :',int(num,8))
    return int(num,8)

def dec_to_hex(num):
    num = int(input('Enter the number : '))
    h = hex(num)
    print('Decimnal to Hexa Value is :',h)
    return h

def hex_to_dec(num):
    num = (input('Enter the number : '))
    int(num,16)
    print('Hexa to Decimal Value is :',int(num,16))
    return int(num,16)

def dec_to_bin(num):
    num = int(input('Enter the number : '))
    b = bin(num)
    print('Decimnal to binary Value is :',b)
    return b

def bin_to_dec(num):
    num = (input('Enter the number : '))
    int(num,2)
    print('Binary to Decimal Value is :',int(num,2))
    return int(num,2)

def char_to_ASCII(char):
    char = input('Enter the number : ')
    asci = ord(char)
    print('Character to ASCII Value is :',asci)
    return asci

def ASCII_to_char(num):
    num = int(input('Enter the number : '))
    char = chr(num)
    print('ASCII to Character Value is :',char)
    return char

def cel_to_far(num):
    num = int(input('Enter the number : '))
    far = num*1.8 + 32
    print('Celcius to Fahrenheit :',far, '°F')
    return far

def far_to_cel(num):
    num = int(input('Enter the number'))
    cel = (num-32)*5/9
    print('Celcius to Fahrenheit :',cel,'°C')
    return cel

def fra_to_dec(n1,n2):
    n1 = int(input('Enter the number'))
    n2 = int(input('Enter the number'))
    dec = n1/n2
    print('Conversion of Fraction to Decimal is :',dec)

```
